```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt

%matplotlib inline

class octahedron_gen(sps.rv_continuous):
    def _pdf(self, x):
        return 5*x**(-6)
```

```python
octahedron= octahedron_gen(a=1, b = 10, name='octahedron')
sample = octahedron.rvs(size = 10000)
#grid = np.linspace(-3, 3, 100)  # задаем сетку для построения графика плотности

plt.figure(figsize=(10,7 ))
plt.xlim(0.8, 5)
plt.ylim(-0.5 , 3)
plt.scatter(sample, np.zeros(10000) , alpha=0.2, label='sample')  # label - описание в легенде
plt.plot(grid, octahedron.pdf(grid), color='red', label='density')  # color - цвет графика
plt.title(r'$p_X(x)= 5x^{-6}$', fontsize = 16)
plt.legend()  # добавляет легенду
plt.show()

dispersion = octahedron.var()
difference = np.zeros(10000)
for n in range(1, 10000):
    dispersion_n = sample[:n].var()
    difference[n] = abs(dispersion- dispersion_n)
plt.figure(figsize=(10,7 ))
plt.ylim(0, 0.1)
plt.plot(range(1, 10000), difference[1:10000], color = 'yellow', label = r'$|S^2(X) - D(X)|$')
plt.legend(fontsize=16)
plt.show()
```
