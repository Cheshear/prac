```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt

%matplotlib inline

M = 100
N = 1000

my_theta=[10, 20, 30]
quadratic_function = [[]]*M

def plot_assay(evaluate , color , label):
    #evaluation=np.zeros(N+1)
    for m in range(M):
        sample = sps.uniform.rvs( loc = 0, scale = theta , size = N)
        #for n in range(N):
        evaluation = np.array([evaluate(sample[:n]) for n in range(1, N + 1)] )
        quadratic_function[m] = (evaluation - theta)**2
    avarage_quadratic_function = [np.mean([quadratic_function[m][n] for m in range(M)])
                                  for n in range(1, N)]
    plt.plot(range(1, N),avarage_quadratic_function, color = color, label = label)
```

```python
theta4 = lambda s: ((s.size+1)/s.size*s.min())
theta3 = lambda s: (s.max() + s.min())
theta2 = lambda s: (s.mean()+s.max())/2
theta1 = lambda s: 2*s.mean()
theta5 = lambda s: (s.size + 1) * s.min()
```

```python
for theta in my_theta:
    plt.figure(figsize=(10, 4))
    plt.ylim(-0.5, 100)
    
    plot_assay(theta1, 'red', r'2$\overline{X}$')
    plot_assay(theta2, 'blue', r'$(\overline{X}+max(X))/2$')
    
    plt.title(r'$theta=$'+str(theta))
    plt.legend(fontsize=16)
    plt.grid(True)
    plt.show()
```

```python
for theta in my_theta:
    plt.figure(figsize=(10, 4))
    plt.ylim(-0.5, 1000)
    
    plot_assay(theta3, 'green', r'$min(X)+max(X)$' )
    plot_assay(theta4, 'pink', r'$\frac{n+1}{n}max(X)$')
    
    plt.title(r'$theta=$'+str(theta))
    plt.legend(fontsize=16)
    plt.grid(True)
    plt.show()
```

```python
for theta in my_theta:
    plt.figure(figsize=(10, 4))
    plt.ylim(-0.5, 1500)
    
    plot_assay(theta5, 'purple', r'$(n+1)min(X)$' )
    
    plt.title(r'$theta=$'+str(theta))
    plt.legend(fontsize=16)
    plt.grid(True)
    plt.show()
```
