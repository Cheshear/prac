```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt

%matplotlib inline

grid = np.linspace(-10, 10, 1000)
arr = [10, 25, 50, 100, 1000, 10000]
r = sps.norm.rvs(loc = 0 ,scale = 1 , size=10000)
for n in arr:
    emp_fun = sps.cumfreq(r[:n], numbins=10000)
    emp_races = emp_fun.lowerlimit + np.linspace(0, emp_fun.binsize * emp_fun.cumcount.size,
                                            emp_fun.cumcount.size)
    
    
    plt.figure(figsize=(10, 4))
    plt.xlim(-2, 2)
    plt.ylim(0, 2)
    plt.step(emp_races, emp_fun.cumcount/n, label = r'$F_n(x)$' )
    plt.plot(grid, sps.norm.cdf(grid), color='green', label=r'$F(x)$')  # color -  
    plt.grid(True)
    plt.legend()  #  
    plt.show()
```

```python
difference = np.zeros(100)
for n in range(1, 100):
    emp_fun = sorted(r[:n])
    emp_races = max(sps.norm.cdf(emp_fun[0]), 1 - sps.norm.cdf(emp_fun[n - 1]))
    
    for i in range(1, n-1):
        emp_races = max(emp_races, abs(sps.norm.cdf(emp_fun[i]) - float(i)/n), abs(sps.norm.cdf(emp_fun[i+1]) - float(i)/n))
    difference[n] =  emp_races   
plt.figure(figsize=(10, 4))
plt.xlim(0, 100)
plt.ylim(-0.2 , 0.2)
plt.plot( range(1,100), difference[1:100],color = 'green',label=r'$\sup_{x\in\mathbb{R}}|F_n(x)-F(x)|$')
plt.legend()
plt.grid(True)
plt.show()
```
