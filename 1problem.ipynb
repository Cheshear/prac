```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
%pylab inline
my_theta=[10, 20, 30]
def plot_of_evaluation(label , appraisal ,color ):
    difference = np.zeros(10000)
    for number in range(1,10000):
        evaluation = appraisal(sample[:number])
        difference[number] = abs(theta - evaluation)
    plt.plot( range(1,10000), difference[1:10000] , color = color, label = label)
```

```python
theta4 = lambda s: ((s.size+1)/s.size*s.min())
third_theta = lambda s: (s.max() + s.min())
theta2 = lambda s: (s.mean()+s.max())/2
theta1 = lambda s: 2*s.mean()
theta5 = lambda s : (s.size + 1) * s.min()
for theta in my_theta:
    plt.figure(figsize=(16, 10))
    plt.xlim(1000, 10000)
    plt.ylim(-10, theta*3)
    sample = sps.uniform.rvs(loc=0, scale=theta, size=10000)
    
    plot_of_evaluation(r'2*$\overline{X}$' ,theta1, 'blue')
    plot_of_evaluation(r'$min(X)+max(X)$' ,theta2, 'red')
    plot_of_evaluation(r'$(\overline{X}+max(X))/2$' ,third_theta, 'yellow')
    plot_of_evaluation(r'$\frac{n+1}{n}max(X)$' ,theta4, 'green')
    plot_of_evaluation(r'$(n+1)min(X)$',theta5 , 'black')
    plt.grid(True)
    plt.title(r'$\theta = '+str(theta))
    plt.legend(fontsize=16)
    plt.show()
```
